```python
correct=0
for i in range(len(x)):
    predict_me=np.array(x[i].astype(float))
    #print(predict_me)
    predict_me=predict_me.reshape(-1,len(predict_me))
    #print(predict_me)
    prediction=model.predict(predict_me)
    if prediction[0] == y[i]:
        correct +=1
print(correct/len(x))        
        
```
